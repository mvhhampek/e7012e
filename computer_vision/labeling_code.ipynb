```python
import os
from PIL import Image
import matplotlib.pyplot as plt
```

```python
dir = "non-labeled_data"
target_dir = "data//Images"
with open("data//labels.txt","a") as label_file:
    for filename in os.listdir(dir):
        f = os.path.join(dir, filename)
        if filename[0:3] == "red":
            label_file.write("\n"+filename+",r")
            os.rename(f, os.path.join(target_dir,filename))
        elif filename[0:4] == "none":
            label_file.write("\n"+filename+",n")
            os.rename(f, os.path.join(target_dir,filename))
        elif filename[0:5] == "green":
            label_file.write("\n"+filename+",g")
            os.rename(f, os.path.join(target_dir,filename))
```

```python
dir = "non-labeled_data"
target_dir = "data//Images"
with open("data//labels.txt","a") as label_file:
    for filename in os.listdir(dir):
        f = os.path.join(dir, filename)
        image = Image.open(f)
        plt.figure()
        plt.imshow(image)
        plt.show()
        inp = input("Write label:")
        if inp not in ('g','r','n'):
            print("Write valid label!")
        else:
            label_file.write("\n"+filename+","+inp)
            os.rename(f, os.path.join(target_dir,filename))
            print(filename, inp)
```
